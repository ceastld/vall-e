```python
from IPython.display import display, Audio, HTML
import subprocess

def infer_aishell2(text_prompts,audio_prompts,output_dir,text) -> str:
    res = subprocess.run(
        [
            "bash",
            "inference.sh",
            "--text-prompts",
            text_prompts,
            "--audio-prompts",
            audio_prompts,
            "--output-dir",
            output_dir,
            "--text",
            text,
        ]
    )
    return text, f'{output_dir}/0.wav'


def infer_aishell2_files(text_prompts_file="prompts/test.txt", use_list=None):
    lines = open(text_prompts_file, "r").readlines()
    
    if use_list:
        lines = [lines[i] for i in use_list]

    for line in lines:
        id = line[:11]
        id1 = line[1:6]
        id2 = line[6:11]
        text_prompts = line[12:].strip()
        audio_prompts = f"download/aishell2/AISHELL-2/iOS/test/wav/{id1}/{id}.wav"
        # audio_prompts = f"prompts/{id}.wav"
        
        display(id)
        
        display_text_audio(text_prompts,audio_prompts)
        
        output_dir = f"infer/{id}"
        text = "我是一头猪"  # 只能用中文标点,英文标点会出问题
        
        display_text_audio(*infer_aishell2(text_prompts,audio_prompts,output_dir,text))

        # print(f'text_prompts="{text_prompts}"\naudio_prompts="./{audio_prompts}"\ntext="{text}"\noutput_dir="{output_dir}"')

infer_aishell2_files("prompts/test.txt")
```
